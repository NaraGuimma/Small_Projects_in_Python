```
idade_pessoas = {
    'Ana' :7,
    'José':38,
    'Paulo':22
}
maiores_de_idade = []
menores_de_idade = []

while len(idade_pessoas) > 0:

  print(len(idade_pessoas))
  pessoas = list(idade_pessoas.keys())
  pessoa = pessoas[0]

  print(pessoas[0])

  if idade_pessoas[pessoa] >=18:
    maiores_de_idade.append(pessoa)

  else:
    menores_de_idade.append(pessoa)

  del idade_pessoas[pessoa]

print(maiores_de_idade)
print(menores_de_idade)
```
